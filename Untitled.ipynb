```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as F

# –°–æ–∑–¥–∞–¥–∏–º —Ç–µ—Å—Ç–æ–≤—ã–π DataFrame
df = spark.range(1, 1001)  # –ß–∏—Å–ª–∞ –æ—Ç 1 –¥–æ 1000

# –ü—Ä–æ—Å—Ç—ã–µ —Ç—Ä–∞–Ω—Å—Ñ–æ—Ä–º–∞—Ü–∏–∏
result = (df
          .withColumn("squared", F.col("id") ** 2)
          .withColumn("is_even", F.col("id") % 2 == 0)
          .filter(F.col("id") <= 10)
         )

print("üìä –ü—Ä–∏–º–µ—Ä —Ä–∞–±–æ—Ç—ã Spark:")
result.show()

print(f"üî¢ –í—Å–µ–≥–æ —Å—Ç—Ä–æ–∫ –≤ –∏—Å—Ö–æ–¥–Ω–æ–º –Ω–∞–±–æ—Ä–µ: {df.count()}")
```
